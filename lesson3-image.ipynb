```python
from PIL import Image

im = Image.open('data/jk.png')
im.size
```

```python
import numpy as np

im_pillow = np.asarray(im)
im_pillow.shape
```

```python
import cv2
im_cv2 = cv2.imread('data/jk.png')
type(im_cv2)
```

```python
im_cv2.shape
```

```python
im_c1 = im_pillow[:,:,0]
im_c2 = im_pillow[:,:,1]
im_c3 = im_pillow[:,:,2]
```

```python
shape = im_pillow.shape[0:2]+(3,)
im_c1_3ch = np.zeros(shape)
im_c2_3ch = np.zeros(shape)
im_c3_3ch = np.zeros(shape)

im_c1_3ch[:,:,0] = im_c1
im_c2_3ch[:,:,1] = im_c2
im_c3_3ch[:,:,2] = im_c3
```

```python
from matplotlib import pyplot as plt
plt.subplot(2, 2, 1)
plt.title('Origin Image')
plt.imshow(im_pillow)
plt.axis('off')
plt.subplot(2, 2, 2)
plt.title('Red Channel')
plt.imshow(im_c1_3ch.astype(np.uint8))
plt.axis('off')
plt.subplot(2, 2, 3)
plt.title('Green Channel')
plt.imshow(im_c2_3ch.astype(np.uint8))
plt.axis('off')
plt.subplot(2, 2, 4)
plt.title('Blue Channel')
plt.imshow(im_c3_3ch.astype(np.uint8))
plt.axis('off')
plt.savefig('./rgb_pillow.png', dpi=150)
```
