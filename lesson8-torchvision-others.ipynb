```python
import torchvision.models as models
import torch

# load pretrained model
googlenet = models.googlenet(weights=models.GoogLeNet_Weights.IMAGENET1K_V1)

fc_in_features = googlenet.fc.in_features
print("fc_in_features:", fc_in_features)

fc_out_features = googlenet.fc.out_features
print("fc_out_features:", fc_out_features)

# change classification number
googlenet.fc = torch.nn.Linear(fc_in_features, 10)
```

```python
import torchvision
from torchvision import datasets
from torchvision import transforms
from torch.utils.data import DataLoader

mnist_data = datasets.MNIST(root='./data',
                            train=True,
                            transform=transforms.ToTensor(),
                            target_transform=None,
                            download=True )

tensor_dataloader = DataLoader(mnist_data, batch_size=32)

img_tensor, label_tensor = next(iter(tensor_dataloader))
print(img_tensor.shape)

grid_tensor = torchvision.utils.make_grid(img_tensor, nrow=8, padding=2)
grid_img = transforms.ToPILImage()(grid_tensor)

display(grid_img)
```
